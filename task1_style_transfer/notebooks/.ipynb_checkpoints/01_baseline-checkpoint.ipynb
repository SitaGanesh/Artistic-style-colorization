```python
"""
01_baseline.ipynb - First Training Notebook for Task 1
Artistic Style Transfer + Colorization
"""

# Add src directory to Python path
import sys
sys.path.append('../src')

# Standard imports
import os
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# Our custom modules
from data_utils import (
    load_image, 
    load_grayscale_image, 
    normalize_for_vgg, 
    lab_to_rgb_tensor, 
    tensor_to_pil,
    get_image_paths
)
from model import StyleColorizer, VGGFeatureExtractor
from losses import compute_content_loss, compute_style_loss, compute_tv_loss

print("All imports successful!")
print(f"PyTorch version: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
```

```python
# Check if we have sample data
content_dir = '../data/content'
style_dir = '../data/styles'

print("Checking data directories...")
print(f"Content dir exists: {os.path.exists(content_dir)}")
print(f"Style dir exists: {os.path.exists(style_dir)}")

if os.path.exists(content_dir):
    content_files = get_image_paths(content_dir)
    print(f"Found {len(content_files)} content images")
    for f in content_files[:3]:  # Show first 3
        print(f"  - {f}")

if os.path.exists(style_dir):
    style_files = get_image_paths(style_dir)
    print(f"Found {len(style_files)} style images")
    for f in style_files[:3]:  # Show first 3
        print(f"  - {f}")
```

```python
# Load one sample of each type
if len(content_files) > 0 and len(style_files) > 0:
    # Load images
    sample_content = content_files[0]
    sample_style = style_files[0]
    
    print(f"Loading content: {sample_content}")
    print(f"Loading style: {sample_style}")
    
    # Load grayscale content
    L_tensor = load_grayscale_image(sample_content, device=device)
    print(f"L tensor shape: {L_tensor.shape}")
    
    # Load style image  
    style_rgb = load_image(sample_style, device=device)
    style_norm = normalize_for_vgg(style_rgb)
    print(f"Style RGB shape: {style_rgb.shape}")
    print(f"Style normalized shape: {style_norm.shape}")
    
    # Visualize loaded images
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    # Show grayscale content
    L_display = L_tensor.squeeze().cpu().numpy()
    axes[0].imshow(L_display, cmap='gray')
    axes[0].set_title('Grayscale Content')
    axes[0].axis('off')
    
    # Show style image
    style_display = style_rgb.squeeze().permute(1,2,0).cpu().numpy()
    axes[1].imshow(style_display)
    axes[1].set_title('Style Reference')
    axes[1].axis('off')
    
    plt.tight_layout()
    plt.show()
    
else:
    print("No sample images found. Add images to data/content/ and data/styles/")
```

```python
# Cell 4: Initialize Model, Test Forward Pass, and Handle Shape Mismatch

print("Initializing StyleColorizer...")
model = StyleColorizer().to(device)

# Avoid re-downloading VGG weights if already cached:
vgg_cache = os.path.expanduser("~/.cache/torch/hub/checkpoints/vgg19-dcbb9e9d.pth")
if not os.path.exists(vgg_cache):
    print("Downloading VGG19 weights (once only)...")
vgg_extractor = VGGFeatureExtractor(
    content_layers=['21'],
    style_layers=['0','5','10','19','28']
).to(device)

print(f"Model parameters: {sum(p.numel() for p in model.parameters()):,}")
print("Model initialized successfully!\n")

# Test forward pass
if len(content_files) > 0 and len(style_files) > 0:
    print("Testing forward pass...")
    model.eval()
    with torch.no_grad():
        ab_pred = model(L_tensor, style_norm)
        # Debug ranges before resizing
        a_chan = ab_pred[0,0]
        b_chan = ab_pred[0,1]
        print(f"DEBUG a*: min {a_chan.min():.3f}, max {a_chan.max():.3f}")
        print(f"DEBUG b*: min {b_chan.min():.3f}, max {b_chan.max():.3f}")
        print(f"Raw ab_pred shape: {ab_pred.shape}")
        print(f"Ab range: [{ab_pred.min().item():.3f}, {ab_pred.max().item():.3f}]")
        
        # Ensure ab_pred matches L_tensor spatial size
        _, _, H, W = L_tensor.shape
        if ab_pred.shape[2:] != (H, W):
            print(f"Resizing ab_pred from {ab_pred.shape[2:]} to {(H, W)}")
            ab_pred = nn.functional.interpolate(ab_pred, size=(H, W), mode='bilinear', align_corners=False)
            print(f"Resized ab_pred shape: {ab_pred.shape}")
        
        # Concatenate and reconstruct
        lab_full = torch.cat([L_tensor, ab_pred], dim=1)
        rgb_recon = lab_to_rgb_tensor(lab_full)
        print(f"Reconstructed RGB shape: {rgb_recon.shape}\n")

    # Visualize initial (untrained) result
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    # Original grayscale
    axes[0].imshow(L_display, cmap='gray')
    axes[0].set_title('Input Grayscale')
    axes[0].axis('off')
    
    # Style reference
    axes[1].imshow(style_display)
    axes[1].set_title('Style Reference')
    axes[1].axis('off')
    
    # Untrained output
    rgb_display = rgb_recon.squeeze().permute(1,2,0).cpu().numpy()
    rgb_display = np.clip(rgb_display, 0, 1)
    axes[2].imshow(rgb_display)
    axes[2].set_title('Untrained Output')
    axes[2].axis('off')
    
    plt.tight_layout()
    plt.show()
else:
    print("No content or style files found; please add sample images to data/content and data/styles.")
```

```python
# Cell 5: Mini Training Loop with Shape and Gradient Handling

# Training hyperparameters
num_epochs = 10
learning_rate = 1e-4
content_weight = 1.0
style_weight = 1e3
tv_weight = 1e-6

# Setup optimizer
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# Training loop
print("Starting mini training loop...")
model.train()

loss_history = []

for epoch in range(num_epochs):
    optimizer.zero_grad()
    
    # Forward pass
    ab_pred = model(L_tensor, style_norm)
    # Apply direct color transfer (optional - for better color matching)
    if len(style_files) > 0:
        from data_utils import get_lab_mean_std
        a_mean, a_std, b_mean, b_std = get_lab_mean_std(style_files[0])
        
        # Re-scale ab_pred
        a = ab_pred[0,0]
        b = ab_pred[0,1]
        a = (a - a.mean()) / (a.std() + 1e-5) * a_std + a_mean
        b = (b - b.mean()) / (b.std() + 1e-5) * b_std + b_mean
        ab_pred = torch.stack([a, b], dim=0).unsqueeze(0)

    # Handle shape mismatch: resize ab_pred to match L_tensor
    _, _, H, W = L_tensor.shape
    if ab_pred.shape[2:] != (H, W):
        ab_pred = nn.functional.interpolate(ab_pred, size=(H, W), mode='bilinear', align_corners=False)
    
    # Reconstruct RGB for feature extraction
    lab_full = torch.cat([L_tensor, ab_pred], dim=1)
    
    # Create a temporary detached version for RGB conversion (to avoid grad issues)
    with torch.no_grad():
        lab_temp = lab_full.detach()
        rgb_pred = lab_to_rgb_tensor(lab_temp)
        rgb_pred_norm = normalize_for_vgg(rgb_pred)
    
    # Re-attach gradients by using the original lab_full for loss computation
    # We'll compute VGG features on the non-grad version but use original tensors for loss
    pred_features = vgg_extractor(rgb_pred_norm)
    style_features = vgg_extractor(style_norm)
    
    # Compute losses using the gradient-enabled tensors
    content_loss = compute_content_loss(
        pred_features['content'], 
        style_features['content']
    )
    
    style_loss = compute_style_loss(
        pred_features['style'], 
        style_features['style']
    )
    
    tv_loss = compute_tv_loss(ab_pred)  # This uses the original ab_pred with gradients
    
    # Total loss
    total_loss = (content_weight * content_loss + 
                  style_weight * style_loss + 
                  tv_weight * tv_loss)
    
    # Backward pass
    total_loss.backward()
    optimizer.step()
    
    # Log progress
    loss_history.append(total_loss.item())
    
    if (epoch + 1) % 2 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}]:')
        print(f'  Total Loss: {total_loss.item():.4f}')
        print(f'  Content: {content_loss.item():.4f}')
        print(f'  Style: {style_loss.item():.6f}') 
        print(f'  TV: {tv_loss.item():.8f}')

print("Mini training completed!")
```

```python
# Cell 6: Visualize Training Progress with Shape Handling

# Plot loss curve
plt.figure(figsize=(10, 6))
plt.plot(loss_history)
plt.title('Training Loss Over Time')
plt.xlabel('Epoch')
plt.ylabel('Total Loss')
plt.grid(True)
plt.show()

# Show final result
print("Generating final result...")
model.eval()
with torch.no_grad():
    ab_final = model(L_tensor, style_norm)
    
    # Handle shape mismatch for final inference too
    _, _, H, W = L_tensor.shape
    if ab_final.shape[2:] != (H, W):
        print(f"Resizing final ab_pred from {ab_final.shape[2:]} to {(H, W)}")
        ab_final = nn.functional.interpolate(ab_final, size=(H, W), mode='bilinear', align_corners=False)
    
    lab_final = torch.cat([L_tensor, ab_final], dim=1)
    rgb_final = lab_to_rgb_tensor(lab_final)

# Before/after comparison
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Input grayscale
axes[0].imshow(L_display, cmap='gray')
axes[0].set_title('Input Grayscale')
axes[0].axis('off')

# Style reference  
axes[1].imshow(style_display)
axes[1].set_title('Style Reference')
axes[1].axis('off')

# Final colorized result
rgb_final_display = rgb_final.squeeze().permute(1,2,0).cpu().numpy()
rgb_final_display = np.clip(rgb_final_display, 0, 1)
axes[2].imshow(rgb_final_display)
axes[2].set_title('Trained Output')
axes[2].axis('off')

plt.tight_layout()
plt.show()
```

```python
# Cell 7: Save Model and Results (no changes needed - uses rgb_final from Cell 6)

# Create output directories
os.makedirs('../models', exist_ok=True)
os.makedirs('../results', exist_ok=True)

# Save trained model
model_path = '../models/baseline_model.pth'
torch.save(model.state_dict(), model_path)
print(f"Model saved to: {model_path}")

# Save result image
result_pil = tensor_to_pil(rgb_final)
result_path = '../results/baseline_result.png'
result_pil.save(result_path)
print(f"Result saved to: {result_path}")

# Save a comparison montage
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
axes[0].imshow(L_display, cmap='gray')
axes[0].set_title('Input')
axes[0].axis('off')

axes[1].imshow(style_display) 
axes[1].set_title('Style')
axes[1].axis('off')

axes[2].imshow(rgb_final_display)
axes[2].set_title('Output')
axes[2].axis('off')

plt.tight_layout()
plt.savefig('../results/baseline_comparison.png', dpi=150, bbox_inches='tight')
plt.show()

print("Baseline training notebook completed successfully!")
```
